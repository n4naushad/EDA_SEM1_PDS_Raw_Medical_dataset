## •	Feature Analysis

* Bar Plot

```python
# Convert columns to numeric, forcing errors to NaN
df_cleaned['Age'] = pd.to_numeric(df_cleaned['Age'], errors='coerce')
df_cleaned['Income'] = pd.to_numeric(df_cleaned['Income'], errors='coerce')

# Bin Age and Income with specified income ranges
df_cleaned['Age_Binned'] = pd.cut(df_cleaned['Age'], bins=[0, 20, 30, 40, 50, 60, 70, 80, 90, 100], right=False)
df_cleaned['Income_Binned'] = pd.cut(df_cleaned['Income'], bins=[0, 30000, 40000, 60000, 80000, 100000], right=False)

# Remove rows with NaN values
df_cleaned.dropna(subset=['Age_Binned', 'Income_Binned'], inplace=True)

# Convert bins to strings for plotting
df_cleaned['Age_Binned'] = df_cleaned['Age_Binned'].astype(str)
df_cleaned['Income_Binned'] = df_cleaned['Income_Binned'].astype(str)

# Create a new DataFrame to count combinations of Age and Income bins
age_income_counts = df_cleaned.groupby(['Age_Binned', 'Income_Binned']).size().reset_index(name='Counts')

# Create a new column for combined categories
age_income_counts['Category'] = age_income_counts['Age_Binned'] + ' & ' + age_income_counts['Income_Binned']

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=age_income_counts, x='Category', y='Counts')
plt.title('Bar Plot of Age and Income')
plt.xlabel('Age and Income')
plt.ylabel('Counts')

# Customize x-axis labels
plt.xticks(ticks=range(len(age_income_counts['Category'])), labels=age_income_counts['Category'].tolist(), rotation=90)

plt.tight_layout()
plt.show()
```
